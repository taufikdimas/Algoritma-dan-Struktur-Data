# D2 Pengenalan Python untuk membaca dan menampilkan image dengan pengenalan macam-macam Library (Numpy, pandas, cv2, skimage, PIL dan matplotlib)

# Langkah 1 : Library


Numpy adalah library manipulasi array, digunakan untuk aljabar linear, transformasi Fourier, dan kemampuan bilangan acak.

Pandas adalah library untuk manipulasi dan analisis data.

CV2 adalah library untuk tugas computer vision.

Skimage adalah library yang mendukung aplikasi pemrosesan citra pada Python.

Matplotlib adalah library yang menghasilkan grafik dan menyediakan toolkit antarmuka pengguna grafis.

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

# Langkah 2 : membaca dan menampilkan image


```python
# Membuat list untuk menyimpan url dari beberapa image
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://yt3.ggpht.com/16mLL14TFnCCw7eLv1Wb2kHYEF2Ce4zfbtmYKrHiVuOEf2HtsSpO3ZAaysQkwcMYyH12HRZvVg=s108-c-k-c0x00ffffff-no-rj"
]

# Baca dan tampilkan image
# Loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)  # read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # resize image to half size
    image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert color to RGB
    final_frame = cv.hconcat([image, image2])  # concatenate image
    cv2_imshow(final_frame)  # show image
    print('\n')
```

# Langkah 3 : melihat ukuran file image

```python
tinggi = image2.shape[0]
lebar = image2.shape[1]
print("Resolusi image: tinggi x lebar = ",tinggi,"x",lebar)
cv2_imshow(image2)
```

# Langkah 4 : mengakses pixel dengan memberikan garis horizontal berwarna putih di tengah image

```python


# Konversi BGR ke RGB
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Buat garis horizontal di tengah gambar
for y in range(lebar):
    image_3[int(tinggi / 2), y] = [255, 255, 255]  # warna putih

# Gabungkan gambar scr horizontal
final_frame = cv.hconcat([image_2, image_3])

cv2_imshow(final_frame)
```
